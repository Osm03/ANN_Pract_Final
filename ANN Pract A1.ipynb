```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Linear Activation Function')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10) 
plt.plot(x, tanh(x)) 
plt.axis('tight') 
plt.title('Tanh Activation Function') 
plt.show() 
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10) 
plt.plot(x, softmax(x)) 
plt.axis('tight') 
plt.title('Softmax Activation Function') 
plt.show() 
```

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10) 
plt.plot(x, RELU(x)) 
plt.axis('tight') 
plt.title('RELU Activation Function') 
plt.show() 
```

```python
def RELU(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-10, 10) 
plt.plot(x, RELU(x)) 
plt.axis('tight') 
plt.title('RELU Activation Function') 
plt.show() 
```
