```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
class1 = np.random.normal(loc=(-1,-1), scale=(0.5,0.5), size=(50,2))
class2 = np.random.normal(loc=(1,1), scale=(0.5,0.5), size=(50,2))
```

```python
X = np.vstack((class1, class2))
y = np.concatenate((np.ones(50), -np.ones(50)))

weights, bias, = np.zeros(X.shape[1]),0
lr, epochs = 0.1, 100
```

```python
for _ in range(epochs):
    for xi, yi in zip(X,y):
        if yi * (np.dot(xi, weights) + bias)<=0:
            weights += lr * yi * xi
            bias += lr * yi
```

```python
x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,0].min() - 1, X[:,0].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

z = np.sign(np.dot(np.c_[xx.ravel(), yy.ravel()], weights) + bias)

z = z.reshape(xx.shape)

plt.contourf(xx, yy, z, alpha=0.2, cmap='PiYG')

plt.scatter(class1[:,0], class1[:,1], color='blue', label='Class 1')
plt.scatter(class2[:,0], class2[:,1], color='red', label='Class 2')

plt.legend()
plt.show()
```
